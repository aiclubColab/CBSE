```python
# List_Operations_Science_Quiz

students = ["Arjun", "Sonakshi", "Vikram", "Sandhya", "Sonal", "Isha", "Kartik"]

# Print whole list
print("Original list:", students)

# Delete "Vikram"
students.remove("Vikram")

# Add "Jay" at the end
students.append("Jay")

# Remove second position (index 1)
students.pop(1)

print("Updated list:", students)
```
