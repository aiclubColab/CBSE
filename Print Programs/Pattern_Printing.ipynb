```python
rows = 5

# Increasing pattern
for i in range(1, rows + 1):
    for j in range(i):
        print("*", end=" ")
    print()

print()

# Decreasing pattern
for i in range(rows, 0, -1):
    for j in range(rows - i):
        print(" ", end=" ")
    for j in range(i):
        print("*", end=" ")
    print()
```
