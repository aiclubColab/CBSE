```python
#Select_Column_and_Display
import pandas as pd

data = {'Name': ['A', 'B', 'C'],
        'Marks': [85, 90, 78]}

df = pd.DataFrame(data)
print(df['Marks'])
```

```python
#Filter_Data_Based_on_Condition
import pandas as pd

data = {'Name': ['A', 'B', 'C'],
        'Marks': [85, 60, 78]}

df = pd.DataFrame(data)
print(df[df['Marks'] > 70])
```

```python
#Add_New_Column_to_DataFrame
import pandas as pd

data = {'Name': ['A', 'B', 'C'],
        'Marks': [85, 90, 78]}

df = pd.DataFrame(data)
df['Grade'] = ['A', 'A', 'B']

print(df)
```
